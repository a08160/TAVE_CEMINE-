```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# 크롬 드라이버 경로 설정
driver = webdriver.Chrome()

# 첫 페이지 URL
url = "https://www.musinsa.com/brands/poloralphlauren?category1DepthCode=001&sortCode=POPULAR_BRAND&page=1&size=120"

# 웹 페이지 열기
driver.get(url)

# 페이지가 완전히 로드될 때까지 기다리기
time.sleep(2)

# 요소들을 담을 리스트
elements = []

# XPath 패턴을 따라 요소들을 찾기
for i in range(1, 121):
    xpath = f'//*[@id="listProducts"]/li[{i}]/div[2]/div[1]/div/strong'
    try:      
        element = driver.find_element(By.XPATH, xpath)
        elements.append(element.text)
    except Exception as e:
        print(f"Element not found for div[{i}]: {e}")

driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/div[8]/div/button[3]').click( )
time.sleep(4)

for i in range(1, 121):
    xpath = f'//*[@id="listProducts"]/li[{i}]/div[2]/div[1]/div/strong'
    try:      
        element = driver.find_element(By.XPATH, xpath)
        elements.append(element.text)
    except Exception as e:
        print(f"Element not found for div[{i}]: {e}")

driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/div[6]/div/button[4]').click( )
time.sleep(4)

for i in range(1, 121):
    xpath = f'//*[@id="listProducts"]/li[{i}]/div[2]/div[1]/div/strong'
    try:      
        element = driver.find_element(By.XPATH, xpath)
        elements.append(element.text)
    except Exception as e:
        print(f"Element not found for div[{i}]: {e}")

driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/div[6]/div/button[5]').click( )
time.sleep(4)

for i in range(1, 121):
    xpath = f'//*[@id="listProducts"]/li[{i}]/div[2]/div[1]/div/strong'
    try:      
        element = driver.find_element(By.XPATH, xpath)
        elements.append(element.text)
    except Exception as e:
        print(f"Element not found for div[{i}]: {e}")

driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/div[6]/div/button[6]').click( )
time.sleep(4)

for i in range(1, 51):
    xpath = f'//*[@id="listProducts"]/li[{i}]/div[2]/div[1]/div/strong'
    try:      
        element = driver.find_element(By.XPATH, xpath)
        elements.append(element.text)
    except Exception as e:
        print(f"Element not found for div[{i}]: {e}")
        

# 요소 내용 출력
for idx, content in enumerate(elements):
    print(f"Element {idx + 1}: {content}")

# 브라우저 닫기
driver.quit()

# elements 리스트를 DataFrame으로 변환
df = pd.DataFrame(elements, columns=["Price"])

# DataFrame을 CSV 파일로 저장
# df.to_csv("calvin전체가격(남성_청바지).csv", index=False, encoding='utf-8-sig')

print(element)
```
